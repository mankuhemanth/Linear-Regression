# Model Creation

```python
# dependent and Independent variable Creation
x=df[["Car_Age_Years","Kms_Driven","Engine_CC","Mileage_kmpl","Num_Owners"]]
y=df["Resale_Value"]
```

```python
x # Displays the Feature Matrics Containing Columns : Car_Age_Years , Kms_Driven , Engine_CC , Mileage_kmpl , Num_Owners
```

```python
y # Displays the Target Variable ( Resale_Value ) as a series
```

```python
# Training and Testing data Creation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
x_train  # Displays the Training Features after train-test-spilt
```

```python
y_train  # Displays the target values( Resale_Value ) for the training set after train-test-spilt
```

```python
# Creating Model
from sklearn.linear_model import LinearRegression  # Importing Linear Regression Model from Scikit-learn
model=LinearRegression()  # Creating Instance( Object ) to the model
```

```python
model.fit(x_train,y_train)  # Training of Linear Regression
```

```python
# Making prediction using x_test to y_test using predict method..
y_pred=model.predict(x_test)  # Predict Resale_values using the trained model on the test set Feature
y_pred 
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Finding the Mean Squared Error(MSE)
mean_squared_error(y_test,y_pred)
```

```python
#Findimng the Root Mean Squared Error (RMSE)
mean_squared_error(y_test,y_pred)**0.5
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score=r2_score(y_test,y_pred)
```
